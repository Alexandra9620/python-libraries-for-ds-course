```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston.keys()
```

```python
X = pd.DataFrame(boston.data, columns = boston.feature_names)
X.info()
```

```python
y = pd.DataFrame(boston.target, columns=["price"])

y.info()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)

y_pred.shape
```

```python
check_test = pd.DataFrame({
    "y_test": y_test["price"],
    "y_pred": y_pred.flatten(),
})

check_test.head()
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error_LR = mean_squared_error(check_test["y_pred"], check_test["y_test"])
print(mean_squared_error_LR)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred_model = model.predict(X_test)
check_test_model = pd.DataFrame({
    "y_test": y_test["price"], 
    "y_pred_model": y_pred_model.flatten()})

check_test_model.head()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error_model = mean_squared_error(check_test_model["y_pred_model"], check_test_model["y_test"])
print(mean_squared_error_model, mean_squared_error_LR)
```

Модель, созданная с помощью RFR работает лучше, чем модель LR

```python
print(model.feature_importances_)
```

```python
feature_importances = pd.DataFrame(model.feature_importances_,
                                   index = X_train.columns,columns=['importance']).sort_values('importance', 
                                   ascending=False)
print(feature_importances)
```
